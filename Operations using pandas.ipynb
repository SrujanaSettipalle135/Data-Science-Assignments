```python
#Step 1: Load the Dataset
import pandas as pd

# Load the dataset
file_path = "/content/diabetes.csv"  # Update the path accordingly
df = pd.read_csv(file_path)
```

```python
#Step 2: Display Basic Information
df.info()  # Shows data types and non-null counts
```

```python
#Step 4: Summary Statistics
df.describe()  # Provides statistical summary (mean, std, min, max, etc.)
```

```python
#Step 5: Check Data Types
df.dtypes  # Shows data types of each column
```

```python
#Step 6: Check for Missing Values
df.isnull().sum()  # Counts missing values in each column
```

```python
#Step 7: Identify Zero Values (Possible Missing Data)
(df == 0).sum()  # Counts zero values in each column
```

```python
#Step 8: Replace Zero Values with NaN
cols_to_replace = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
df[cols_to_replace] = df[cols_to_replace].replace(0, pd.NA)
```

```python
#Step 9: Fill Missing Values with the Median
df.fillna(df.median(), inplace=True)
```

```python
#Step 10: Correlation Analysis
df.corr()  # Computes pairwise correlation between features
```

```python
#Step 11: Group Data by Diabetes Outcome
df.groupby("Outcome").mean()  # Shows mean statistics for diabetic and non-diabetic groups
```

```python
#Step 12: Creating a New Feature – BMI Category
df["BMI_Category"] = pd.cut(df["BMI"], bins=[0, 18.5, 24.9, 29.9, 40],
                            labels=["Underweight", "Normal", "Overweight", "Obese"])
```

```python
#Step 13: Sorting Data
df.sort_values(by=["Age", "BMI"], ascending=[True, False])  # Sort by Age (ascending) and BMI (descending)
```

```python
#Step 14: Select a Single Column
df["Age"]  # Selects only the Age column
```

```python
#Step 15: Select Multiple Columns
df[["Glucose", "BMI"]]  # Selects Glucose and BMI columns
```

```python
#Step 16: Find the Top 5 Oldest Diabetic Individuals
df[df["Outcome"] == 1].nlargest(5, "Age")
```
